# 0 PREPARATION

```python
# Packages
import numpy as np
import pandas as pd
```

```python
# Options
pd.set_option('display.max_colwidth', 200)
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 20)
pd.options.display.float_format = "{:.2f}".format
```

```python
# Constants
YEAR_START=2014
YEAR_END=2023
```

----
# 1 DATA EXTRACTION

```python
# Load data
years=np.arange(YEAR_START,YEAR_END+1,1)
dfs = []
for year in years:
  url = f"https://raw.githubusercontent.com/JeffSackmann/tennis_wta/master/wta_matches_{str(year)}.csv"
  print(url)
  df_tmp = pd.read_csv(url)
  df_tmp['year'] = year
  dfs.append(df_tmp)
df_raw = pd.concat(dfs, ignore_index=True)
df_raw
```
